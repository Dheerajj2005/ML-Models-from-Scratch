# **Understanding Train Test Split**

## Intuition
When building any machine learning model, we want it to **generalize** —  
not just perform well on the data it’s trained on, but also on **new unseen data**.

If we train and test on the **same dataset**, we might get misleadingly high accuracy,  
because the model has already seen those samples. This is known as **overfitting**.

To fix this, we split our dataset into two parts:
- **Training set:** Used to train the model.
- **Testing set:** Used to evaluate the model’s performance on unseen data.

---

##  Why Do We Need It?

Let’s say we are predicting whether a student passes or fails based on study hours.

If we train on all 100 students and then test on those same 100,  
our model might achieve 100% accuracy — but it tells us nothing about how it will perform on *new students*.

Train-Test Split helps simulate the real-world scenario:
> "Train on past data → Predict on future unseen data."

---

##  Typical Split Ratios

| Ratio | Meaning |
|--------|----------|
| 80% / 20% | Common for medium-sized datasets |
| 70% / 30% | Safer for smaller datasets |
| 60% / 40% | Used for high-variance models |

---

##  Formula / Concept

If our dataset has `N` samples:
- Training data = `N × (train_size)`
- Testing data = `N × (1 - train_size)`

For example, with 1000 samples and train_size = 0.8:  
→ 800 for training, 200 for testing.

---

##  Workflow

1. Load the dataset  
2. Split into features (X) and labels (y)  
3. Divide into training and test sets  
4. Train a simple model  
5. Evaluate performance on both sets

# **Dataset Preparation**

```python
# Create a simple dataset
np.random.seed(42)
X = np.random.randn(20, 2)
y = np.random.choice([0, 1], size=20)

df = pd.DataFrame(np.hstack((X, y.reshape(-1, 1))), columns=['Feature1', 'Feature2', 'Label'])
df.head()
```

## **Implementation**

Custom implementation of train-test split.
    
    Parameters:
    -----------
    X : np.ndarray
        Feature matrix (inputs)
    y : np.ndarray
        Target vector (labels)
    test_size : float, default=0.2
        Fraction of dataset to use for testing (e.g., 0.2 = 20% test data)
    shuffle : bool, default=True
        Whether to randomly shuffle the data before splitting
    random_state : int, optional
        Seed for reproducibility when shuffling
    
    Returns:
    --------
    X_train, X_test, y_train, y_test : np.ndarray
        Splits of the input data

```python
def train_test_split_custom(X, y, test_size=0.2, shuffle=True, random_state=None):
    # if shuffle is True randomize the order of data points
    if shuffle:
        # set a random seed so it genearate the same split every time
        if random_state is not None:
            np.random.seed(random_state)

        # shuffle dataset indices
        indices = np.random.permutation(len(X))
    else:
        # no shuffling, use ordered indices (0, 1, 2, ...)
        indices = np.arange(len(X))

    # no of samples for test set
    test_count = int(len(X) * test_size)

    # Split indices into test and train parts
    test_indices = indices[:test_count]
    train_indices = indices[test_count:]

    # Use the indices to pick rows from X and y
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]

    return X_train, X_test, y_train, y_test
```

```python
# Implemented Function
X_train, X_test, y_train, y_test = train_test_split_custom(X, y, test_size=0.3, random_state=42)

print("Training set shape:", X_train.shape)
print("Testing set shape:\n", X_test.shape)

print("Sample training data:\n", X_train[:5])
print("Sample test data:\n", X_test[:5])
```

```python
# Compare with scikit learn's train_test_split
from sklearn.model_selection import train_test_split

X_train_lib, X_test_lib, y_train_lib, y_test_lib = train_test_split(
    X, y, test_size=0.3, random_state=42, shuffle=True
)

print("Scikit-learn Split Shapes:")
print("Train:", X_train_lib.shape, " Test:", X_test_lib.shape)
```

```python
# Are Both Splits Similar
print("Custom Train Mean:", np.mean(X_train))
print("Sklearn Train Mean:", np.mean(X_train_lib))
print("Difference in mean:", abs(np.mean(X_train) - np.mean(X_train_lib)))
```
