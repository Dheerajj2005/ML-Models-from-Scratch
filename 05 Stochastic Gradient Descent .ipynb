## Stochastic Gradient Descent 

```python
import pandas as pd
import numpy as np
```

**Problem with Batch Gradient Descent/Gradient Descent for m Dimension**
1) It's computation cost is very high so it make training very slow for large datasets <br>
2) In batch gradient descent we perform vectorization to calculate y-hat for this operation we have to load the whole dataset in RAM if the data is too big then you can't perform this operation <br>

**Stochastic Gradient Descent (SGD)** <br>
Instead of looking at the entire dataset for each adjustment, SGD takes a shortcut. It picks a single random data point (or a very small batch) from the dataset. It then calculates how to adjust the model's settings based on just that one point.

**Advantage** <br>
1) Faster Conveergence <br>


**Disadvantage** <br>
1) Solution is not steady <br>

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score 
import time
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

### Simple Linear Regression

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test, y_pred)
```

### Built Model of SGD

```python
class SGDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                instance = np.random.randint(0,X_train.shape[0])
                
                y_hat = np.dot(X_train[instance],self.coef_) + self.intercept_
                
                intercept_der = -2 * (y_train[instance] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                
                coef_der = -2 * np.dot((y_train[instance] - y_hat),X_train[instance])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
sgd = SGDRegressor(learning_rate=0.01, epochs=40)
```

```python
start = time.time()
sgd.fit(X_train, y_train)
print("The time taken is : ", time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
